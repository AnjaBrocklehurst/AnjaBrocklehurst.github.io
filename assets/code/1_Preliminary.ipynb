```python
#Import pandas, files and look at the data (show first two rown and ask for info)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set pandas display options to show all columns for .head command
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)       # Auto-detect the display width
pd.set_option('display.max_colwidth', None) # Show full content of each column

# Loading the datasets
file_path1 = r'ByCompany.csv'
Company_data = pd.read_csv(file_path1, sep=';')

file_path2 = r'ByTargets.csv'
Target_data = pd.read_csv(file_path2, sep=';')
```

```python
# Assuming Company_data is already loaded
# Count occurrences of each sector
sector_counts = Company_data['sector'].value_counts().reset_index()
sector_counts.columns = ['sector', 'count']

# Create the barplot
plt.figure(figsize=(10, 6))
sns.barplot(data=sector_counts, x='sector', y='count', palette='viridis')

# Customize the plot
plt.title('Count of Occurrences by Sector', fontsize=16)
plt.xlabel('Sector', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
